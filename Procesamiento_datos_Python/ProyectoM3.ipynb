 ## Conexión directa con Mongo DB

Se requiere la instalación de pandas `!pip install pymongo`

Para este ejemplo se conectara a servidor local, base de datos **Proyecto2**

```python
from pymongo import MongoClient
from bson.json_util import dumps, loads

client = pymongo.MongoClient('localhost',27017)
#Lista bases de datos de servidor
client.list_database_names()
```

```python
#Utilizar base de datos proyecto2
db = client['proyecto2']
#Coleciones
db.list_collection_names()
```

```python
collection = db['ventas']
#Numero de documentos
collection.count_documents({})
```

```python
#Ver informacion de cursor 
info = collection.find()
#for cr in info:
#    print(cr)
```

```python
#Convertir cursor a Json
cursor = collection.find()
list_cur = list(cursor)
json_data = dumps(list_cur)
#json_data
```

```python
#Convertir cursor a Dataframe
df = pd.DataFrame(list_cur)
df
```

## Conexión API local 

Se requiere la instalación de `!pip install requests`

Para este ejemplo se conectara a servidor local, base de datos **http://localhost:3000/api/craigslists**

```python
import requests
endpoint = 'http://localhost:3000/api/craigslists'
r = requests.get(endpoint)
```

```python
r.json().keys()
```

```python
json = r.json()
result = json['craigslist']
```

> Convertir resultado obtenido en el API a Dataframe

```python
normalized = pd.json_normalize(result)

df = pd.DataFrame.from_dict(normalized)

df.head()
```

> Es posible acceder a esta información leyendo la propiedad shape de nuestro DataFrame mostrando numero de registros y columnas:

```python
df.shape
```

```python
#Trae registro especifico
df.loc[5]
```

```python
#Trae un rango de registros especificos
df.loc[5:7]
```

```python
#Elimina columnas con informacion no relevante
df=df.drop(columns=['_id','id','url','region_url','imagen_url','fechapublicacion'])
#Primeras filas 
df.head()
```

```python
#Ultimas Filas
df.tail()
```

```python
#La propiedad dtypes nos da una Serie donde podemos ver cada columna y el tipo de dato que contiene:
df.dtypes
```

```python
#Contar cuántos NaNs existen usando la agregación sum, que suma 1 por cada True y 0 por cada False
df.isna().sum(axis=0)
```

```python
#Con axis=0 nos regresa el conteo por columnas. Con axis=1 nos regresa el conteo por filas:
df.isna().sum(axis=1)
```

Para limpiar las filas que tengan mínimo 1 valor NaN `df.dropna(axis=0, how='any')`
>Cambiar valores NAN por un valor

```python
df['tamanio'] = df['tamanio'].fillna('sin tamanio')
df.head()
```

```python
#Cambia el NAN en manejo por default awd
df['manejo'] = df['manejo'].fillna('awd')
df.head()
```
